# Population Growth Rate Formula
$\text{Growth Rate} = \frac{\text{Present}-\text{Past}}{\text{Past}}$
<br>
bacteria in [One Hour] has Growth Rate %10
$\text{Growth Rate } = \frac{110-100}{100}$
<br>
$\text{Growth Rate = 0.1}$
<br>
(new population) = (old population) + (change in population)
<br>
or
<br>
population(t) = population($t -\Delta t$) + $\Delta\text{population}$
<br>
$\frac{dP}{dt} =rP$
<br>
<br>
The constant r is the growth rate, or instantaneous growth rate, or continuous
growth rate, while dP/dt is the rate of change of the population.
<br>
**Algorithm 1 Algorithm for simulation of unconstrained growth**
<pre>
    initialize simulationLength
    initialize population
    initialize growthRate
    initialize length of time step ∆t
    numIterations ← simulationLength/∆t
    for i going from 1 through numIterations do the following:
        growth ← growthRate * population
        population ← population + growth * ∆t
        t ← i * ∆t
        display t, growth, and population
</pre>

```python
bacteria_population=100
growth_rate=0.1
delta_time=0.005
time_total=0
bacteria_population_record=[]
time_total_record=[]
```

```python
def bacteria_population_function(bacteria_population=100,
                                 growth_rate=0.1,
                                 delta_time=0.005,
                                 time_total=0,
                                 stop_time=1):
    
    global bacteria_population_record
    global time_total_record
    bacteria_population_record=[]
    time_total_record=[]
    while(time_total<stop_time):
        bacteria_population=bacteria_population+(growth_rate*bacteria_population)*delta_time
        bacteria_population_record.append(bacteria_population)
        time_total=time_total+delta_time
        time_total_record.append(time_total)
    print(bacteria_population)
```

```python
def bacteria_population_graph():
    %matplotlib inline
    import matplotlib.pyplot as plt
    fig,ax = plt.subplots()
    ax.plot(time_total_record,bacteria_population_record)
```

# Another way to calculate is using the for loop
**Algorithm 2 Alternative algorithm to Algorithm 1 for simulation of unconstrained growth that does not display growth**
<pre>
initialize simulationLength
initialize population
initialize growthRate
initialize ∆t
growthRatePerStep ← growthRate * ∆t
numIterations ← simulationLength/∆t
for i going from 1 through numIterations do the following:
    population ← population + growthRatePerStep * population
    t ← i * ∆t
    display t and population
</pre>
$\large \text{Number of iterations loop}=\frac{\text{Time}}{\text{Accuracy}(\Delta{t})}$

```python
bacteria_population=100
growth_rate=0.1
delta_time=0.005
time_total=0
bacteria_population_record=[]
time_total_record=[]

stop_time=30
loop_iterations=int(stop_time/delta_time)
```

```python
for alpha in range(loop_iterations):
    bacteria_population=bacteria_population+growth_rate*bacteria_population*delta_time
    bacteria_population_record.append(bacteria_population)
    time_total=time_total+delta_time
    time_total_record.append(time_total)
#     print(f"{alpha:<5}|{time_total:^10.3f}|{bacteria_population:^10.3f}")
```

# Carrying Capacity $M$
$\LARGE\frac{dP}{dt}=r \Bigl(1-\frac{P}{M}\Bigl) P$
<br>
$\frac{dN}{dt}=\text{rate of population change}$
<br>
 $ r=\text{maximum population growth rate}$
<br>
$ N=\text{population size}$
<br>
$ K=\text{population carrying capacity}$


```python
def bacteria_population_function_carrying_capacity(bacteria_population=100,
                                 growth_rate=0.1,
                                 delta_time=0.00005,
                                 time_total=0,
                                 stop_time=100,
                                  M=1000):
    global bacteria_population_record
    global time_total_record
    bacteria_population_record=[]
    time_total_record=[]
    while(time_total<stop_time):
#         bacteria_population=(bacteria_population+bacteria_population*growth_rate)*(1-bacteria_population*delta_time/M)
        bacteria_population=bacteria_population+((bacteria_population*growth_rate)-(growth_rate*bacteria_population/M)*bacteria_population)*delta_time
        bacteria_population_record.append(bacteria_population)
        time_total=time_total+delta_time
        time_total_record.append(time_total)
    print(bacteria_population)
bacteria_population_function_carrying_capacity()
bacteria_population_graph()
```

```python
def plot_accuracy_scatter(expected,predicted):
    import matplotlib.pyplot as plt
    import matplotlib.patches as mpatches
    # Creates a figure
    fig,ax = plt.subplots()
    x=np.arange(max(max(predicted),max(expected)))
    ax.plot(x,x,color='darkorange',linestyle="--")
    target_line = mpatches.Patch(color='darkorange', label='Target Line')
    ax.legend(handles=[target_line])

    ax.scatter(expected,predicted)
    plt.xlabel('Expected')
    plt.ylabel('Predicted')
    plt.show()
    
```

```python
def scatter_matrix_custom(data_frame,attributes,fig_size=(50,50)):
    %matplotlib inline
    fig,ax=plt.subplots(nrows=len(attributes)-1, ncols=len(attributes)-1,figsize=fig_size)
    for i in range(len(attributes)-1):
        for j in range(len(attributes)-1):
            if(i==j):
                ax[i,j].hist(data_frame[attributes[i]])
            else:
                ax[i,j].scatter(data_frame[attributes[j]],data_frame[attributes[i]])
            if(i==len(attributes)-2):
                ax[i,j].set(xlabel=attributes[j])
            if(j==0):
                ax[i,j].set(ylabel=attributes[i])
#     plt.savefig(f'scatter_matrix_custom.png',dpi=700)
```
